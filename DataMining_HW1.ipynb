# First Attempt at Ex 2.2 on pg 39

I couldn't get the boundary lines to match exactly like the example in the textbook.

```python
# Parameters
meta_mean_blue = [1, 0]
meta_mean_orange = [0, 1]
meta_covariance = np.eye(2)
low_variance_covariance = np.eye(2) / 5
num_samples = 10000
num_means = 10

# Generate means for each class
means_blue = np.random.multivariate_normal(meta_mean_blue, meta_covariance, num_means)
means_orange = np.random.multivariate_normal(meta_mean_orange, meta_covariance, num_means)

# Generate samples
def generate_samples(means, num_samples):
    chosen_means = means[np.random.choice(num_means, num_samples)]
    return np.array([np.random.multivariate_normal(mean, low_variance_covariance) for mean in chosen_means])

samples_blue = generate_samples(means_blue, num_samples)
samples_orange = generate_samples(means_orange, num_samples)
```

```python
from scipy.stats import multivariate_normal

def posterior_probability(sample, means):
    probabilities = [multivariate_normal.pdf(sample, mean=mean, cov=low_variance_covariance) for mean in means]
    return sum(probabilities) / len(means)

# Evaluate posterior probabilities over a grid
x = np.linspace(samples_blue[:,0].min() - 1, samples_blue[:,0].max() + 1, 200)
y = np.linspace(samples_blue[:,1].min() - 1, samples_blue[:,1].max() + 1, 200)
X, Y = np.meshgrid(x, y)
Z_blue = np.vectorize(lambda x,y: posterior_probability([x,y], means_blue))(X, Y)
Z_orange = np.vectorize(lambda x,y: posterior_probability([x,y], means_orange))(X, Y)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(samples_blue[:, 0], samples_blue[:, 1], c='blue', alpha=0.5, label='Blue')
plt.scatter(samples_orange[:, 0], samples_orange[:, 1], c='orange', alpha=0.5, label='Orange')
plt.contour(X, Y, Z_blue - Z_orange, levels=[0], colors='black')
plt.legend()
plt.title("Bayes Decision Boundary")
plt.show()
```

# Second Attempt at Ex. 2.2 on pg 39

```python
# Generate 10 random mean values
blue = np.random.multivariate_normal([1,0], np.eye(2), 10)
orange = np.random.multivariate_normal([0,1], np.eye(2), 10)

def create_observations(prob_class):
  observations = []
  for _ in range(100):
      # Randomly select a mean from random_means
      mk = prob_class[np.random.choice(10)]

      # Generate a data point using the selected mean
      observation = np.random.multivariate_normal(mk, np.identity(2)/5)
      observations.append(observation)

  return np.array(observations)

blue_observations = create_observations(blue)
orange_observations = create_observations(orange)
```

```python
from sklearn.svm import SVC

# Combine blue and orange data points
X = np.vstack((blue_observations, orange_observations))
y = np.array([0]*len(blue_observations) + [1]*len(orange_observations))  # 0 for blue, 1 for orange

# Train SVM with RBF kernel
clf = SVC(kernel='rbf', C=1E6)
clf.fit(X, y)

# Plot data points
plt.scatter(blue_observations[:, 0], blue_observations[:, 1], color='blue', label='Blue Data', alpha=0.5)
plt.scatter(orange_observations[:, 0], orange_observations[:, 1], color='orange', label='Orange Data', alpha=0.5)

# Plot decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# Plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[0], linestyles=['-'])

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Scatter Plot with Decision Boundary')
plt.show()
```
